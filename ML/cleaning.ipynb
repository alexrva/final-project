```python
housingTime = []
for row in range(0, len(housing)):
    for j in range(7, len(headers)):
        oldrec = housing.iloc[row]
        newrec = {'RegionID': oldrec['RegionID'], 'RegionName': oldrec['RegionName'], 'Statecode': oldrec['State'], 'Date': headers[j] + "-01", 'Price': oldrec[headers[j]]}
        housingTime.append(newrec)
    
```

```python
import csv
with open('housing1.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['RegionID','RegionName','Statecode','Date','Price'])
    writer.writeheader()
    writer.writerows(housingTime)
```
